```python
import glob
import json
import pandas as pd
from tabulate import tabulate
from textwrap import wrap
```

```python
context_title = "José Luis Picardo"

with open(f"indices/{context_title}/docstore.json", "r") as f:
    original_docstore = json.load(f)
with open(f"indices/{context_title}_summaries/docstore.json", "r") as f:
    summary_docstore = json.load(f)
```

```python
original_chunks = [value["__data__"]["text"] for key, value in original_docstore["docstore/data"].items()]
summaries = [value["__data__"]["text"] for key, value in summary_docstore["docstore/data"].items()]
```

```python
# Create table comparing each original_chunks[i] with summaries[i]
table = []
for i in range(len(original_chunks)):
    table.append(["\n".join(wrap(original_chunks[i])), "\n".join(wrap(summaries[i]))])

df = pd.DataFrame(table, columns=["Original", "Summary"])
# Print table with lines between rows
print(tabulate(df, headers="keys", tablefmt="grid"))
```
