```python
import os


def get_subdirectories_without_underscore(directory):
    # List all entries in the directory
    entries = os.listdir(directory)
    # Filter entries to find directories without an underscore
    return [entry for entry in entries if "_" not in entry and os.path.isdir(os.path.join(directory, entry))]


# Specify the directory path
directory_path = r"C:\AbhiData\Desktop\UPennCourses\CIS 6200\Project\LongRAG\indices"

# Call the function and print the result
files_without_underscore = get_subdirectories_without_underscore(directory_path)
print(files_without_underscore)
```
